## **Part 3: Reflection**

*In a brief paragraph, compare the two maps you created. Describe the visual and data model differences between the DEM datasets visualized in the topology map and the zonal statistics calculated and appended to the parcel dataset in the second map.*

<mark> First off, the Oxford Topography map or (Map1) is based on a Raster dataset that highlights the change in elevation, going pixel by pixel. This gives a continous look, as each pixel blends smoothly next to one another. Next, the Oxford Slope Topography map is a choropleth map that is vector based as it uses non-continous visuals to show parcel by parcel differences in mean slope. This is a result of the first map using a DEM and the second using a zonal statistics combination of a parcel layer and the appended mean slope differences layer resulting in a completely opposite data model. By using the zonal statistics tool, a calculation of a single mean slope value from the underlying slope raster for each parcel polygon was made, which was then averaged across the entire terrain in each specific parcel. </mark>
